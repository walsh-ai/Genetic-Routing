**Erdos-Renyi Random Graph**
The Erdos-Renyi random graph model generates random graphs by one of two methods. Either, a random graph G(n, M) is produced having n nodes M edges, chosen uniformly from the set of all graphs having n nodes and M edges. Or else, a random graph G(n, p) is generated, where each pair of nodes is connected with probability p independently of all other pairs. 

```python
import random 
import itertools
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt 
```

```python
def erdos_renyi(n, p):
    # Create an empty networkx graph
    G = nx.Graph()
    
    # Add n nodes
    G.add_nodes_from(range(n))
    
    # Add possible edges with probability p
    # Disallow self-edges
    for i in range(n):
        for j in range(n):
            if i == j:
                continue 
            
            rnd = random.random()
            
            if rnd < p:
                G.add_edge(i, j)
    
    return G
```

```python
G = erdos_renyi(100, 0.05)
nx.draw(G)
```
