```python
# Redo this from scratch with the same graphs 
# for both diameter and clustering 
# Then, make the two separate plots and plot them against one another 
# on a logarithmic scale 
# Scale the diameters to be between zero and one 

# STEP 1: Produce probability range 
ps = np.append(np.linspace(0, 0.05, 8), np.linspace(0.1, 1, 18))

# STEP 2: Produce the graphs 
graphs = [] 
for p in ps:
    graphs.append(nx.watts_strogatz_graph(1000, 10, p))

# STEP 3: Calculate diameter and global clustering 
gc = []
diam = [] 

for g in graphs:
    gc.append(nx.transitivity(g))
    diam.append(nx.diameter(g))
    
# STEP 4: Produce scaled diam 
# Normalized data= ( data- min(data) )/( max(data)-min(data) )
diam_ = (diam - np.min(diam)) / (np.max(diam) - np.min(diam))
```

```python
# STEP 5: Plot the gc 
f = plt.figure() 
f.set_size_inches(8, 6)
plt.title("Global Clustering vs. Randomness")
plt.scatter(ps, gc, label="Transitivity")
plt.legend() 
plt.show()

f.savefig("clustering.jpg", dpi=100)
```

```python
# STEP 6: Plot gc logarithmic 
fig = plt.figure()
fig.set_size_inches(8, 6)
gcx = plt.gca()
plt.title("Global Clustering vs. Randomness (Log)")
gcx.scatter(ps, gc, label="Transitivity")
gcx.set_yscale('log')
gcx.legend()

fig.savefig("clustering_log.jpg", dpi=100)
```

```python
# STEP 7: Plot the diam logarithmic 
fig = plt.figure()
gcx = plt.gca()
fig.set_size_inches(8, 6)
plt.title("Characteristic Path Length vs. Randomness (Log)")
gcx.scatter(ps, diam, label="Diameter")
gcx.set_yscale('log')
gcx.legend()



fig.savefig("diameter_log.jpg", dpi=100)
```

```python
# STEP 8: Plot gc and diam_ together 
fig = plt.figure()
gcx = plt.gca()
fig.set_size_inches(8, 6)
plt.title("Clustering and Diameter vs. Randomness")
gcx.scatter(ps, gc, label="Transitivity")
gcx.scatter(ps, diam_, label="Diameter")
gcx.set_yscale('log')
gcx.set_xscale('log')
gcx.legend()



fig.savefig("gc_diam.jpg", dpi=100)
```

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

#G = nx.gnp_random_graph(100, 0.02, seed=10374196)
G = nx.barabasi_albert_graph(n = 50, m = 2, seed=10374196, initial_graph = None)

degree_sequence = sorted((d for n, d in G.degree()), reverse=True)
dmax = max(degree_sequence)

fig = plt.figure("Degree of a random graph", figsize=(8, 8))
# Create a gridspec for adding subplots of different sizes
axgrid = fig.add_gridspec(5, 4)

ax0 = fig.add_subplot(axgrid[0:3, :])
Gcc = G.subgraph(sorted(nx.connected_components(G), key=len, reverse=True)[0])
pos = nx.spring_layout(Gcc, seed=10396953)
nx.draw_networkx_nodes(Gcc, pos, ax=ax0, node_size=20)
nx.draw_networkx_edges(Gcc, pos, ax=ax0, alpha=0.4)
ax0.set_title("Scale Free Networks: Being Lonely is Normal")
ax0.set_axis_off()

ax1 = fig.add_subplot(axgrid[3:, :2])
ax1.plot(degree_sequence, "b-", marker="o")
ax1.set_title("Degree Rank Plot")
ax1.set_ylabel("Degree")
ax1.set_xlabel("Rank")

ax2 = fig.add_subplot(axgrid[3:, 2:])
ax2.bar(*np.unique(degree_sequence, return_counts=True))
ax2.set_title("Degree histogram")
ax2.set_xlabel("Degree")
ax2.set_ylabel("# of Nodes")

fig.tight_layout()
plt.show()

fig.savefig("scale_free.jpg")
```
